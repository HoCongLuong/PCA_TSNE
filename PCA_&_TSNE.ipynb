# Thuật toán PCA:

```
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

keys = np.array(feature_df.keys()[3:])
values = feature_df[keys].values
values = StandardScaler().fit_transform(values)

pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(values)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

principalDf.plot.scatter(x = 'PC1', y = 'PC2')
```

# Thuật toán TSNE


```
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

data = feature_df.loc[:, 'danceability':]

standardized_data = StandardScaler().fit_transform(data)

model = TSNE(n_components = 2, random_state = 0)
tsne_data = model.fit_transform(data)

# Tạo ra 1 dataframe cho việc trực quan kết quả
tsne_df = pd.DataFrame(data = tsne_data,
     columns =("Dim_1", "Dim_2"))

tsne_df.plot.scatter(x = 'Dim_1', y = 'Dim_2')
```
